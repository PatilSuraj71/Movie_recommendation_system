```python
import pandas as pd
import numpy as np
import ast
from sklearn.feature_extraction.text import CountVectorizer
import nltk
import pickle
import requests

```

```python
credit=pd.read_csv(r"tmdb_5000_credits.csv")
```

```python
movies=pd.read_csv(r"tmdb_5000_movies.csv")
```

```python
movies.head()
```

```python
movies1=movies.merge(credit,on='title')
```

```python
movies1.head()
```

```python
movies1=movies1[["id","title",'overview','genres','keywords','cast','crew']]
```

```python
movies1.head()
```

```python
movies1.shape
```

```python
movies1.isnull().sum()
```

```python
movies1.dropna(inplace=True)
```

```python
movies1.duplicated().sum()
```

```python
movies1.iloc[0].genres
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies1["genres"]=movies1["genres"].apply(convert)
```

```python
movies1.head()
```

```python
movies1["keywords"]=movies1["keywords"].apply(convert)
```

```python
movies1["cast"].iloc[0]
```

```python
def convert1(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
           l.append(i['name'])
           counter=counter+1
        else:
            break
    return l

```

```python
movies1["cast"]=movies1["cast"].apply(convert1)
```

```python
movies1['crew'].iloc[0]
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            l.append(i['name'])
            break
    return l
```

```python
movies1["crew"]=movies1["crew"].apply(fetch_director)
```

```python
movies1.head()
```

```python
movies1["overview"].iloc[0]
```

```python
movies1["overview"]=movies1["overview"].apply(lambda x: x.split())
```

```python
movies1["genres"]=movies1["genres"].apply(lambda x:[i.replace (" ","") for i in x])
```

```python
movies1["keywords"]=movies1["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies1["tags"]=movies1["overview"]+movies1["keywords"]+movies1["cast"]+movies1["crew"]
```

```python
new_movies=movies1[["id","title","tags"]]
```

```python
new_movies.head()
```

```python
new_movies["tags"]=new_movies["tags"].apply(lambda x:" ".join(x))
```

```python
new_movies["tags"].iloc[0]
```

```python
new_movies["tags"]=new_movies["tags"].apply(lambda x:x.lower())
```

```python
le=CountVectorizer(max_features=5000,stop_words='english')
vectors=le.fit_transform(new_movies["tags"]).toarray()

```

```python
vectors[0]
```

```python
le.get_feature_names_out()

```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loving')
```

```python
new_movies['tags']=new_movies["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities=cosine_similarity(vectors)
```

```python
similarities[0]
```

#enumerate=touple from


```python
sorted(list(enumerate(similarities[0])), reverse=True,key=lambda x:x[1])[1:6]

```

```python
new_movies[new_movies["title"]=='Spectre'].index[0]
```

```python
def recommend_movies(movie):
    movie_index = new_movies[new_movies["title"] == movie].index[0]

    distance = similarities[movie_index]

    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend_movies('Avatar')
```
